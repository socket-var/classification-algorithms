```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
train_file = "train_features.csv"

data = pd.read_csv(train_file, header=None)

X_train = data.iloc[:, 1:].copy()

labels_file = "train_label.csv"
data = pd.read_csv(labels_file)
y_train = data.iloc[:, -1].copy()

test_file = "test_features.csv"

data = pd.read_csv(test_file, header=None)

X_test = data.iloc[:, 1:].copy()


print(X_train.head(), y_train.head(), X_test.head())
```

```python
# KNN
neigh = KNeighborsClassifier()
neigh.fit(X_train, y_train) 
```

```python
neigh.predict(X_test)
```

```python
# Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
gnb.predict(X_test)
```

```python
# Decision tree classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
# random forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)  
```

```python
clf.predict(X_test)
```
